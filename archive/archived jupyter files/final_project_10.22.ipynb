# Factor Analysis

```python
from sklearn.decomposition import FactorAnalysis
X, y = total_df.drop(['diagnosis'], axis =1), total_df.diagnosis
factor = FactorAnalysis(n_components=30, random_state=7).fit(X)
factor.get_params()
```

```python
factor.components_
```

```python
factor_df = pd.DataFrame(factor.components_, columns=columns)
```

```python
factor_df.max()
```

```python
factor_df.max().plot(kind='bar')

plt.title('Factor Analysis Bar Chart')

plt.savefig('total_df_factor_bar.png')
```

# Logistic Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
